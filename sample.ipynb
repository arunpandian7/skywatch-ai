```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
def plot_images(img1, img2):
    plt.figure(figsize=(10,10))

    img1 = Image.open(img1)
    img2 = Image.open(img2)

    ax1 = plt.subplot(1,2,1)
    ax1.imshow(img1)

    ax2 = plt.subplot(1,2,2)
    ax2.imshow(img2)

    plt.show()
```

```python
from skywatchai.SkywatchAI import compare

test1 = 'test/image1.jpg'
test2 = 'test/image2.jpg'

plot_images(test1, test2)
print('Are they same person:',compare(test1, test2))
```

```python
test3 = 'test/image3.jpg'

plot_images(test2, test3)

print('Are they same person:', compare(test2, test3))
```
